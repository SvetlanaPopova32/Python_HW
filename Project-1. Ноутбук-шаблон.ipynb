Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

```python
#ваш код здесь
# Функция для преобразования значения столбца
def get_education(row):
    words = row.split()  # Разбиваем строку на отдельные слова
    education_words = []
    for word in words:
        # Если слово состоит только из букв и не является числом, добавляем его в список
        if word.isalpha() and not word.isdigit():
            education_words.append(word)
        else:
            # Как только встречаем числовое значение (дату), останавливаемся
            break
    education_string = ' '.join(education_words)  # Преобразуем список слов в строку
    return education_string

# Создаем новый столбец 'Образование'
data['Образование'] = data['Образование и ВУЗ'].apply(get_education)

# Проверяем полученные категории
categories = data['Образование'].unique()
print(categories)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
def extract_work_experience(experience_str):
  """
  Функция извлекает из строки "Опыт работы" информацию о 
  общем опыте работы в месяцах.

  Args:
    experience_str: строка, содержащая информацию о опыте работы

  Returns:
    int or NaN: общее количество месяцев опыта работы или NaN, 
    если информация отсутствует или некорректна
  """

  if experience_str is None or str(experience_str) == "Не указано":
    return np.nan

  # Разбиваем строку на слова
  words = str(experience_str).split()


  # Переменные для хранения значений лет и месяцев
  years = 0
  months = 0

  # Итерируемся по словам
  for i in range(len(words)-1):
    if words[i].isdigit():
      # Если слово является числом, то оно может обозначать 
      # либо количество лет, либо количество месяцев. смотрим по следующему слову
      if (words[i + 1] == "год" or words[i + 1] == "года" or words[i + 1] == "лет"):
        years = int(words[i])
        if words[i+2].isdigit() == False:
          return years * 12
      

      elif (words[i + 1] == "месяц" or words[i + 1] == "месяца" or words[i + 1] == "месяцев"):
        months = int(words[i])
        if months !=0:
          return years * 12 + months


  # Возвращаем общее количество месяцев опыта работы

data['Опыт работы (месяц)'] = data['Опыт работы'].apply(extract_work_experience)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
```

```python
def transform_city_info(city_str):
  """
  Функция преобразует информацию о городе в три отдельных признака.

  Args:
    city_str: строка, содержащая информацию о городе, переезде и командировках

  Returns:
    str, bool, bool: город, готовность к переезду, готовность к командировкам
  """

  
  city = None
  relocation = False
  commissions = False

  # Разбиваем строку на слова
  words = city_str.split(" , ")

  # Ищем информацию о городе
  city = words[0]

  # Ищем информацию о командировках
  for word in words:
    if "командировкам" in word:
      if "не" in word:
        commissions = False
      else:
        commissions = True


  # Ищем информацию о переезде
  for word in words:
    if "переезду" in word:
      if "не" not in word:
        relocation = True
    elif "хочу" in word:
      relocation = True

  # Определяем категорию города
  if city in million_cities:
    city = "город-миллионник"
  elif city in ["Москва", "Санкт-Петербург"]:
    pass
  else:
    city = "другие"

  return city, relocation, commissions


# Применяем функцию к каждому значению в столбце "Город, переезд, командировки"
for i in range(len(data)):
  city_info = data.loc[i, "Город, переезд, командировки"]
  city, relocation, commissions = transform_city_info(city_info)

  data.loc[i, "Город"] = city
  data.loc[i, "Готовность к переезду"] = relocation
  data.loc[i, "Готовность к командировкам"] = commissions
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
# ### Категории занятости и графика

employment_categories = ["частичная занятость", "проектная работа", "полная занятость", "стажировка", "волонтерство", "вахтовый метод"]
schedule_categories = ["полный день", "сменный график", "гибкий график", "удаленная работа", "вахтовый метод"]



# ### One-Hot Encoding с помощью цикла

for category in employment_categories:
  data[category] = data["Занятость"].apply(lambda x: category in x.split(", "))

for category in schedule_categories:
  data[category] = data["График"].apply(lambda x: category in x.split(", "))
  
data.drop(['Занятость', 'График'], axis=1, inplace=True)
```

```python
#ваш код здесь

# Преобразование "Обновление резюме"
data["date"] = pd.to_datetime(data["Обновление резюме"], format='mixed')
data["date"] = data["date"].dt.strftime("%Y.%m.%d")

# Преобразование "date" в таблице с курсами
df_currencies["date"] = pd.to_datetime(df_currencies["date"], format='mixed')

# Extract just the year, month, and day
df_currencies["date"] = df_currencies["date"].dt.strftime("%Y.%m.%d")

# Извлечение информации из "ЗП"
data["amount"] = data["ЗП"].str.split().str[0]
data["amount"] = data["amount"].apply(lambda x: int(x))
data["currency"] = data["ЗП"].str.split().str[1]

# Словарь соответствия названий валют и их кодов
currency_codes = {
  "руб.": "RUB",
  "бел.руб.": "BYN",
  "грн.": "UAH",
  "сум": "UZS",
  "KZT": "KZT",
  "KGS": "KGS",
  "AZN": "AZN",
  "EUR": "EUR",
  "USD": "USD"
}
data["currency"] = data["currency"].apply(lambda x: currency_codes.get(x))

# Объединение данных
df_merged = pd.merge(data, df_currencies, how="left", on=["date", "currency"])

# Заполнение close для RUB (assuming the function is for filling)
def fill_close_rub(df):
    df.loc[df["currency"] == "RUB", "close"] = df["close"].fillna(1)
    df.loc[df["currency"] == "RUB", "proportion"] = df["proportion"].fillna(1)
    return df

df_merged = fill_close_rub(df_merged.copy())  # Apply the function to a copy


df_merged["ЗП (руб)"] = df_merged["amount"] * df_merged["close"] / df_merged["proportion"]
```

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# ваш код здесь
# Логарифмирование возраста
log_age = np.log(df_cleaned['Возраст'])

# Распределение в логарифмическом масштабе
sns.histplot(data=log_age, stat='density', log_scale=True)

# Среднее значение
mean = log_age.mean()


# Boundaries (3 sigma with 1 sigma slack to the right)
sigma = log_age.std()
lower_bound = mean - 3 * sigma
upper_bound = mean + 4 * sigma

# Vertical lines for mean and upper bound
plt.axvline(mean, color='k', lw=2, label='Среднее')
plt.axvline(upper_bound, color='r', lw=2, label='4σ')
plt.axvline(lower_bound, color='r', lw=2, label='3σ')

# Legend
plt.legend()
plt.show()
```

```python
def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned


outliers, cleaned = outliers_z_score(df_cleaned, 'Возраст', log_scale=True)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
```
